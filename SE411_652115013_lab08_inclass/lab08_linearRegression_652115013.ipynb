```python
# Read and show data table
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

data = pd.read_csv('cleaned_all_phones.csv')
data.describe()
```

```python
# Plotting price distribution in graph
y_label = 'price(USD)' # predicts
x1_label = 'inches' # constraints
y = data[y_label]
x = data[x1_label]
plt.scatter(x,y)
plt.ylabel(y_label, fontsize = 20)
plt.xlabel(x1_label, fontsize = 20)
plt.show()
```

```python
x_const = sm.add_constant(x)
results = sm.OLS(y,x_const).fit()
results.summary()
```

```python
# Plotting the Regression line
plt.scatter(x,y)
yhat = 107.7878*x + (-354.4156)
fig = plt.plot(x, yhat, lw=4, c='orange', label = 'regression line')
plt.xlabel(x1_label, fontsize = 20)
plt.ylabel(y_label, fontsize = 20)
plt.show()
```
